```python
#!/user/bin/env python
# -*- coding: utf-8 -*-
from requests_oauthlib import OAuth1Session
import csv
import json
import sys, codecs



search_words = u"山中俊治"

C_KEY = 'CVV3JwFouRIbMtUy6YuK6uFd2'
C_SECRET = '6Ymtz3G8CQQBi00lOX3HTQeq2L7Bo8ry8T3kaJvXYSDvbogEGH'
A_TOKEN = '3223473522-ZDwrX0N0cOyCr66yzowhUADHAOy0YLSn8yjp2bs'
A_TOKEN_SECRET = '73cnEPm9J85yIzhamFPdw10C26f9UGRN8R9gIE1otsF4a'



def Search_words():
    url = "https://api.twitter.com/1.1/search/tweets.json?"
    params = {
            "q": search_words,
            "lang": "ja",
            "result_type": "recent",
            "count": "100"
            }
    tw = OAuth1Session(C_KEY,C_SECRET,A_KEY,A_SECRET)
    req = tw.get(url, params = params)
    tweets = json.loads(req.text)

    f = open("tweetsearch.csv" , "a")
    writer = csv.writer(f)
    writer.writerow(["datetime", "id", "name", "text"])

    for tweet in tweets["statuses"]:
        time = (tweet["created_at"])
        id = (tweet["user"]["screen_name"])
        name = (tweet["user"]["name"])
        text = (tweet["text"])

        writer.writerow([time, id, name, text])

    f.close()

    return Search_words

Search_words()
```
